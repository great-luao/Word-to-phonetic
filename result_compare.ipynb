```python
# remove the space in a certain file
def remove_space(file):
    with open(file, 'r') as f:
        lines = f.readlines()
    with open(file, 'w') as f:
            for line in lines:
                line = line.replace(' ', '')
                f.write(line)

# Iterate all files in the yrcy_temp folder
import os
for root, dirs, files in os.walk('../yrcy/temp'):
    for file in files:
        if file.endswith('.txt'):
            remove_space(os.path.join(root, file))
```

```python
# Compare the same name txt files in the yrcy_temp and yrcy folders
# Find all words, check if they are the same
# If not, store the different words in a dictionary
# Don't store the same different words again.
import difflib
import re
diff_words = {}
same_wrods = []
for root, dirs, files in os.walk('../yrcy/temp'):
    for file in files:
        if file.endswith('.txt'):
            with open(os.path.join(root, file), 'r', encoding='utf8') as f1:
                lines1 = f1.readlines()
            with open(os.path.join('../yrcy', file), 'r', encoding='utf8') as f2:
                lines2 = f2.readlines()
            for line1, line2 in zip(lines1, lines2):
                parts1 = re.findall(r'\[.*?\]|\(.*?\)|\w+|[^\s\w]', line1)
                parts2 = re.findall(r'\[.*?\]|\(.*?\)|\w+|[^\s\w]', line2)
                if(len(parts1) != len(parts2)):
                    print(f'{file} has different length of lines')
                    print(parts1)
                    print(parts2)
                    continue
                for word1, word2 in zip(parts1, parts2):
                    if not re.match (r'\w+', word1) or not re.match(r'\w+', word2): continue 
                    if word1 != word2:
                        if word1 not in diff_words:
                            diff_words[word1] = word2
                    else:
                        if word1 not in same_wrods:
                            same_wrods.append(word1)

print(diff_words)
print(same_wrods)
print(f'The misatake rate is {len(diff_words) / (len(diff_words) + len(same_wrods)):.2f}')
```

```python
# Write the different words to a file
with open('diff_words.txt', 'w', encoding='utf8') as f:
    for k, v in diff_words.items():
        f.write(f'{k} -> {v}\n')
```
